```python
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")
```

```python
image_path1 = "/content/sample_data/B.png"
base64_image1 = encode_image(image_path1)
print(base64_image1)
image_path2 = "/content/sample_data/NISI20111218_0005614773_web.jpg"
base64_image2 = encode_image(image_path2)
print(base64_image2)
image_path3 = "/content/sample_data/eccf287206c6df94f412fca502e6555f63638b06edb07a8571e8fbf0e10f.jpg"
base64_image3 = encode_image(image_path3)
print(base64_image3)
```
